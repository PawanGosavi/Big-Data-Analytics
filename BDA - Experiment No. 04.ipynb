```python
from pymongo import *
import pymongo
```

```python
client = pymongo.MongoClient("mongodb://localhost:27017/")
```

```python
db = client["college_data"]
```

```python
my_collection = db["student_data"]
```

```python
student_record = {
   "Name": "Pawan Gosavi",
   "Age": 23,
   "Sex": "M",
   "Roll No.": 1,
   "Class": "MScIT P1"
}
```

```python
my_collection.insert_one(student_record)
```

```python
for item in my_collection.find():
    print(item)
```
